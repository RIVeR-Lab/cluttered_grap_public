```python
rolling_imgs = []
NUM_TO_AVERAGE = 3

for i in range(9):
    rolling_imgs.append(cv.imread("d" + str(i+1) + ".png"))
    
for img in rolling_imgs:
    plt.figure()
    plt.imshow(img)
```

```python

mega_img = np.zeros(rolling_imgs[0].shape)
for img in rolling_imgs:
    mega_img += (img / len(rolling_imgs))
    plt.figure()
    plt.imshow(cv.convertScaleAbs(mega_img))
```
