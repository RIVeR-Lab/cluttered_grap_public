```python
import math
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv
import random
# Authored by Gary Lvov and Nathaniel Hanson
```

```python
img = cv.imread('overhead.png')
bbx = [0, img.shape[1]]
bby = [0, img.shape[0]]

PRIMITIVES_RADIUS_SIZE = 75
PRIMITIVES_ANGLE_VAR = .314
```

```python
x = np.linspace(bbx[0],bbx[1],100)
y = np.linspace(bby[0], bby[1], 100)


def plot_bbx():
    plt.imshow(cv.imread('overhead.png'))
    # for line in bbx:
    #     plt.plot(np.ones(100) * line, y, color="red")

    # for line in bby:
    #     plt.plot(x, np.ones(100) * line, color="red")

    # plt.plot(0, 0, 'ro')
plt.show()
```

```python
plot_bbx()
```

```python
def create_potential_moves(x, y):
    plt.figure(figsize=(10,10))
    plt.plot(x, y, 'ro')
    plt.axis("equal")
    plt.axis('off')
    plot_bbx()
    
    c = plt.Circle((x, y), radius = PRIMITIVES_RADIUS_SIZE, alpha = .1)
    plt.gca().add_artist(c)
    potential_moves = []
    
    for angle in range(1, 11):
        x_offset = PRIMITIVES_RADIUS_SIZE * math.sin(angle * PRIMITIVES_ANGLE_VAR)
        y_offset = PRIMITIVES_RADIUS_SIZE * math.cos(angle * PRIMITIVES_ANGLE_VAR)
        move = [x - x_offset, y - y_offset, 
                x + x_offset, y + y_offset]
        potential_moves.append(move)
        #plt.plot(move[0], move[1], move[2], move[3], linewidth=3)
        plt.plot(x - x_offset, y - y_offset, 'ro', color="blue", markersize=5)
        plt.plot(x + x_offset, y + y_offset, 'ro', color="red", markersize=5)
    return potential_moves
plt.show()
```

```python
moves = create_potential_moves(100, 100)

for x in range(5):
    x = random.randint(0, img.shape[1])
    y = random.randint(0, img.shape[0])
    moves = create_potential_moves(x, y)
    plt.show()
```
